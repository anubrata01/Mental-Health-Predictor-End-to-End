#### Creating new columns to structure the dataset.

```python
social_media_columns = ["facebook","instagram","youtube","snapchat","reddit","discord","pinterest","tiktok"]
```

```python
print([i.strip().lower() for i in df["which social media platforms"][0].split(",")])
```

```python
df[social_media_columns]=0
```

- initialized all the new columns with 0

```python
df.head()
```

- We'll use lambda function

```python
for platform in social_media_columns:
    df[platform]= df["which social media platforms"].apply(
        lambda x:1 if platform in [i.strip().lower() for i in x.split(",")] else 0
    )
```

```python
df.head()
```

- social media count

```python
df["social_media_count"] = df[social_media_columns].sum(axis=1)
```

```python
df["social_media_count"].describe()
```
