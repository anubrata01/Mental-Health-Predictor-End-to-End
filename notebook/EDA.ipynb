#### Checking for co relation

```python
columns_for_co_relation=['using Social media without a specific purpose','distracted by Social media while working','feel restless if haven\'t used Social media in a while'
,'how easily distracted are you','how much are you bothered by worries',
'find it difficult to concentrate on things','how often do you compare yourself',
'how do you feel about these comparisons','seek validation from features of social media',
'feel depressed or down','how frequently interest fluctuate daily',
'sleep issue',
'facebook','instagram','youtube','snapchat','reddit','discord','pinterest','tiktok']
```

```python
corr = df[columns_for_co_relation].corr()
plt.figure(figsize=(18,8))
sns.heatmap(corr,annot=True,cmap='coolwarm',fmt='.2f',linewidths=0.5)
plt.title("Correlation Heatmap of Selected Social Media & Mental Health Factors")
plt.show()
```

#### Which columns to drop?

You drop columns based on two reasons:

a. High inter-correlation (multicollinearity):

If two variables have correlation > 0.85 (either + or âˆ’), they carry almost the same information.
â†’ Keep one, drop the other.

Example:

If â€˜how easily distracted are youâ€™ and â€˜distracted by social media while workingâ€™ = 0.90
â†’ Drop one of them.

b. Very weak correlation with target variable:

If your main prediction target is, say, â€˜feel depressed or downâ€™,
then you can check its correlation with all other features:

corr['feel depressed or down'].sort_values(ascending=False)


Drop features with correlation between âˆ’0.1 and +0.1, since they add little predictive power.

```python
corr_target = corr['feel depressed or down'].sort_values(ascending=False)

plt.figure(figsize=(8,10))
sns.heatmap(corr_target.to_frame(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
# .to_frame() turns a series into dataframe
plt.title("Correlation Heatmap with Target Variable ('feel depressed or down')")
plt.show() 
```
